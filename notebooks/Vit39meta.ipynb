```python
# ==========================================
# ĐÁNH GIÁ PHÂN LOẠI 9 LỚP KẾT HỢP 3 LỚP ENSEMBLE (REFINE)
# ==========================================
# 1. Import thư viện và thiết lập môi trường
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import random
import cv2
from sklearn.metrics import confusion_matrix, classification_report, balanced_accuracy_score, roc_auc_score, accuracy_score

import torch
import torch.nn as nn
import torch.nn.functional as F

import timm
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

import albumentations as A
from albumentations.pytorch import ToTensorV2

# Seed để tái tạo kết quả
def seed_everything(seed=42):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

seed_everything()

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Sử dụng device: {device}')

# 2. Đường dẫn dữ liệu (giống NineClasses)
SPLIT_DIR = '/kaggle/input/isic-stest'
TRAIN_IMG_DIR = os.path.join(SPLIT_DIR, 'train/images')
TRAIN_SEG_DIR = os.path.join(SPLIT_DIR, 'train/segmentations')
TRAIN_LABELS_PATH = os.path.join(SPLIT_DIR, 'train_labels.csv')
TRAIN_METADATA_PATH = os.path.join(SPLIT_DIR, 'train_metadata.csv')
TEST_IMG_DIR = os.path.join(SPLIT_DIR, 'valid/images')
TEST_SEG_DIR = os.path.join(SPLIT_DIR, 'valid/segmentations')
TEST_LABELS_PATH = os.path.join(SPLIT_DIR, 'valid_labels.csv')
TEST_METADATA_PATH = os.path.join(SPLIT_DIR, 'valid_metadata.csv')
VAL_IMG_DIR = os.path.join(SPLIT_DIR, 'test/images')
VAL_SEG_DIR = os.path.join(SPLIT_DIR, 'test/segmentations')
VAL_LABELS_PATH = os.path.join(SPLIT_DIR, 'test_labels.csv')
VAL_METADATA_PATH = os.path.join(SPLIT_DIR, 'test_metadata.csv')

os.makedirs('results', exist_ok=True)

# 3. Hàm chuyển đổi nhãn one-hot sang multiclass
def convert_to_multiclass(df):
    label_columns = [col for col in df.columns if col not in ['image', 'diagnosis', 'label']]
    result_df = df.copy()
    result_df['diagnosis'] = result_df[label_columns].idxmax(axis=1)
    class_names = sorted(label_columns)
    label_map = {class_name: i for i, class_name in enumerate(class_names)}
    result_df['label'] = result_df['diagnosis'].map(label_map)
    return result_df, label_map, class_names

# 4. Xử lý metadata
def process_metadata(metadata_df):
    metadata_df = metadata_df[['image', 'age_approx', 'anatom_site_general', 'sex']].copy()
    metadata_df['age_approx'] = pd.to_numeric(metadata_df['age_approx'], errors='coerce')
    age_mean = metadata_df['age_approx'].mean()
    metadata_df['age_approx'] = metadata_df['age_approx'].fillna(age_mean)
    age_min = metadata_df['age_approx'].min()
    age_max = metadata_df['age_approx'].max()
    metadata_df['age_approx'] = (metadata_df['age_approx'] - age_min) / (age_max - age_min)
    metadata_df['anatom_site_general'] = metadata_df['anatom_site_general'].fillna('unknown')
    site_dummies = pd.get_dummies(metadata_df['anatom_site_general'], prefix='site')
    metadata_df['sex'] = metadata_df['sex'].fillna('unknown')
    sex_dummies = pd.get_dummies(metadata_df['sex'], prefix='sex')
    metadata_processed = pd.concat([
        metadata_df[['image']],
        metadata_df[['age_approx']],
        site_dummies,
        sex_dummies
    ], axis=1)
    metadata_processed = metadata_processed.fillna(0)
    meta_feature_cols = metadata_processed.columns.tolist()[1:]
    return metadata_processed, meta_feature_cols

# 5. Augmentation
def get_valid_transforms(use_seg=True):
    height, width = 224, 224
    n_channels = 3 + (1 if use_seg else 0)
    return A.Compose([
        A.Resize(height=height, width=width),
        A.Normalize(mean=[0.485, 0.456, 0.406] + [0.5] * (n_channels - 3),
                   std=[0.229, 0.224, 0.225] + [0.5] * (n_channels - 3)),
        ToTensorV2(),
    ])

# 6. Dataset class
class DermDataset(torch.utils.data.Dataset):
    def __init__(self, img_dir, df, metadata_df=None, meta_cols=None, seg_dir=None, 
                 transform=None, use_seg=True, use_meta=False):
        self.img_dir = img_dir
        self.df = df
        self.metadata_df = metadata_df
        self.meta_cols = meta_cols
        self.seg_dir = seg_dir
        self.transform = transform
        self.use_seg = use_seg
        self.use_meta = use_meta
        self.image_names = df['image'].values
        self.labels = df['label'].values

    def __len__(self):
        return len(self.df)
    
    def __getitem__(self, idx):
        img_name = self.image_names[idx]
        img_path = os.path.join(self.img_dir, img_name + '.jpg')
        img = cv2.imread(img_path)
        if img is None:
            img_path = os.path.join(self.img_dir, img_name + '.jpeg')
            img = cv2.imread(img_path)
        if img is None:
            img_path = os.path.join(self.img_dir, img_name + '.png')
            img = cv2.imread(img_path)
        if img is None:
            img = np.zeros((224, 224, 3), dtype=np.uint8)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        if self.use_seg and self.seg_dir is not None:
            seg_path = os.path.join(self.seg_dir, img_name + '_seg.png')
            if os.path.exists(seg_path):
                seg_mask = cv2.imread(seg_path, cv2.IMREAD_GRAYSCALE)
                if seg_mask.shape[:2] != img.shape[:2]:
                    seg_mask = cv2.resize(seg_mask, (img.shape[1], img.shape[0]), interpolation=cv2.INTER_NEAREST)
                if seg_mask.max() > 1:
                    seg_mask = seg_mask / 255.0
                img = np.dstack((img, seg_mask))
        if self.transform:
            augmented = self.transform(image=img)
            image = augmented['image']
        else:
            image = img
        meta_features = None
        if self.use_meta and self.metadata_df is not None and self.meta_cols is not None:
            meta_row = self.metadata_df[self.metadata_df['image'] == img_name]
            if not meta_row.empty:
                try:
                    meta_values = meta_row[self.meta_cols].values[0].astype(np.float32)
                    meta_features = torch.tensor(meta_values, dtype=torch.float32)
                except Exception:
                    meta_features = torch.zeros(len(self.meta_cols), dtype=torch.float32)
            else:
                meta_features = torch.zeros(len(self.meta_cols), dtype=torch.float32)
        label = torch.tensor(self.labels[idx], dtype=torch.long)
        if meta_features is not None:
            return image, meta_features, label
        else:
            return image, label

# 7. Model class (ViT/EfficientNet with meta)
class DermModelWithMeta(nn.Module):
    def __init__(self, model_name='vit_base_patch16_224', num_classes=9, use_seg=True, 
                use_meta=True, meta_dim=0, pretrained=True):
        super(DermModelWithMeta, self).__init__()
        self.use_seg = use_seg
        self.use_meta = use_meta
        in_channels = 3 + (1 if use_seg else 0)
        self.backbone = timm.create_model(
            model_name, 
            pretrained=pretrained,
            in_chans=in_channels
        )
        if 'vit' in model_name or 'deit' in model_name or 'swin' in model_name:
            backbone_out = self.backbone.head.in_features
            self.backbone.head = nn.Identity()
        elif 'efficientnet' in model_name:
            backbone_out = self.backbone.classifier.in_features
            self.backbone.classifier = nn.Identity()
        elif 'resnet' in model_name:
            backbone_out = self.backbone.fc.in_features
            self.backbone.fc = nn.Identity()
        else:
            raise NotImplementedError(f"Model {model_name} chưa được hỗ trợ")
        self.meta_dim = meta_dim
        if use_meta and meta_dim > 0:
            self.meta_processor = nn.Sequential(
                nn.Linear(meta_dim, 128),
                nn.ReLU(),
                nn.Dropout(0.3),
                nn.Linear(128, 64),
                nn.ReLU(),
                nn.Dropout(0.2)
            )
            self.classifier = nn.Sequential(
                nn.Linear(backbone_out + 64, 512),
                nn.ReLU(),
                nn.Dropout(0.3),
                nn.Linear(512, num_classes)
            )
        else:
            self.meta_processor = None
            self.classifier = nn.Linear(backbone_out, num_classes)
    def forward(self, x, meta=None):
        img_features = self.backbone(x)
        if self.use_meta and meta is not None and self.meta_processor is not None:
            meta_features = self.meta_processor(meta)
            combined_features = torch.cat([img_features, meta_features], dim=1)
            output = self.classifier(combined_features)
        else:
            output = self.classifier(img_features)
        return output

# 8. StackingEnsemble và VotingEnsemble (từ BKLens)
class StackingEnsemble(nn.Module):
    def __init__(self, base_models, num_classes=3, meta_dim=0):
        super(StackingEnsemble, self).__init__()
        self.base_models = nn.ModuleList(base_models)
        self.num_classes = num_classes
        meta_learner_input = len(base_models) * num_classes
        if meta_dim > 0:
            self.meta_processor = nn.Sequential(
                nn.Linear(meta_dim, 32),
                nn.ReLU(),
                nn.Dropout(0.2),
            )
            meta_learner_input += 32
        else:
            self.meta_processor = None
        self.meta_learner = nn.Sequential(
            nn.Linear(meta_learner_input, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(32, num_classes)
        )
    def forward(self, x, meta=None):
        base_preds = []
        for model in self.base_models:
            model.eval()
            with torch.no_grad():
                if meta is not None and hasattr(model, 'meta_processor') and model.meta_processor is not None:
                    outputs = model(x, meta)
                else:
                    outputs = model(x)
                probs = F.softmax(outputs, dim=1)
                base_preds.append(probs)
        ensemble_features = torch.cat(base_preds, dim=1)
        if meta is not None and self.meta_processor is not None:
            meta_features = self.meta_processor(meta)
            ensemble_features = torch.cat([ensemble_features, meta_features], dim=1)
        output = self.meta_learner(ensemble_features)
        return output

class VotingEnsemble(nn.Module):
    def __init__(self, base_models, weights=None):
        super(VotingEnsemble, self).__init__()
        self.base_models = nn.ModuleList(base_models)
        if weights is None:
            self.weights = torch.ones(len(base_models)) / len(base_models)
        else:
            self.weights = weights
    def forward(self, x, meta=None):
        all_outputs = []
        for i, model in enumerate(self.base_models):
            model.eval()
            with torch.no_grad():
                if meta is not None and hasattr(model, 'meta_processor') and model.meta_processor is not None:
                    output = model(x, meta)
                else:
                    output = model(x)
                all_outputs.append(F.softmax(output, dim=1) * self.weights[i])
        final_output = sum(all_outputs)
        return final_output

# 9. Hàm load checkpoint hoặc full model thông minh
def load_model_smart(model_path, device, model_architecture='vit_base_patch16_224', num_classes=3, meta_dim=0):
    print(f"Đang tải mô hình từ {model_path}...")
    try:
        model = torch.load(model_path, map_location=device, weights_only=False)
        if isinstance(model, nn.Module):
            print(f"✓ Đã tải full model thành công")
            return model.to(device)
        if isinstance(model, dict):
            print(f"✓ Đã tải checkpoint thành công, đang khởi tạo mô hình...")
            new_model = DermModelWithMeta(
                model_name=model_architecture,
                num_classes=num_classes,
                use_seg=True,
                use_meta=True,
                meta_dim=meta_dim,
                pretrained=False
            )
            if 'model_state_dict' in model:
                new_model.load_state_dict(model['model_state_dict'])
                print(f"✓ Đã tải state_dict từ checkpoint")
            elif 'state_dict' in model:
                new_model.load_state_dict(model['state_dict'])
                print(f"✓ Đã tải state_dict từ checkpoint")
            else:
                try:
                    new_model.load_state_dict(model)
                    print(f"✓ Đã tải state_dict trực tiếp")
                except:
                    print(f"⚠️ Không tìm thấy state_dict trong checkpoint")
            return new_model.to(device)
        print(f"⚠️ Định dạng model không xác định, thử tạo mô hình mới...")
        new_model = DermModelWithMeta(
            model_name=model_architecture,
            num_classes=num_classes,
            use_seg=True,
            use_meta=True,
            meta_dim=meta_dim,
            pretrained=True
        )
        return new_model.to(device)
    except Exception as e:
        print(f"❌ Lỗi khi tải model: {e}")

# 10. Chuẩn bị dữ liệu test
print("Đọc dữ liệu test...")
test_df = pd.read_csv(TEST_LABELS_PATH)
test_df, label_map, class_names = convert_to_multiclass(test_df)
num_classes = len(class_names)
print(f"Số lớp: {num_classes}, class_names: {class_names}")

metadata_df = pd.concat([
    pd.read_csv(TRAIN_METADATA_PATH),
    pd.read_csv(VAL_METADATA_PATH),
    pd.read_csv(TEST_METADATA_PATH)
])
metadata_processed, meta_feature_cols = process_metadata(metadata_df)

test_dataset = DermDataset(
    img_dir=TEST_IMG_DIR,
    df=test_df,
    metadata_df=metadata_processed,
    meta_cols=meta_feature_cols,
    seg_dir=TEST_SEG_DIR,
    transform=get_valid_transforms(use_seg=True),
    use_seg=True,
    use_meta=True
)
test_loader = torch.utils.data.DataLoader(
    test_dataset, batch_size=32, shuffle=False, num_workers=2, pin_memory=True
)

# 11. Load các mô hình
multiclass_model_path = "/kaggle/input/tokenmixsmallval/pytorch/default/5/checkpoints/vit_meta_multiclass_tokenmix_best.pth"
vit_model_path = '/kaggle/input/bklnvmel/pytorch/default/1/checkpoints/vit_base_patch16_224_best.pth'
eff_model_path = '/kaggle/input/bklnvmel/pytorch/default/2/checkpoints/vit_meta_multiclass_focal_best.pth'
three_class_model_path = '/kaggle/input/bklnvmel/pytorch/default/3/stacking_ensemble_model.pth'

print("Tải mô hình 9 lớp...")
model_9class = load_model_smart(
    multiclass_model_path, device,
    model_architecture='vit_base_patch16_224',
    num_classes=9, meta_dim=len(meta_feature_cols)
)
print("Tải mô hình ViT 3 lớp...")
model_vit3 = load_model_smart(
    vit_model_path, device,
    model_architecture='vit_base_patch16_224',
    num_classes=3, meta_dim=len(meta_feature_cols)
)
print("Tải mô hình EfficientNet 3 lớp...")
model_eff3 = load_model_smart(
    eff_model_path, device,
    model_architecture='efficientnet_b5',
    num_classes=3, meta_dim=len(meta_feature_cols)
)
print("Tải mô hình ensemble 3 lớp...")
model_ensemble3 = torch.load(three_class_model_path, map_location=device, weights_only=False)
model_ensemble3 = model_ensemble3.to(device)
model_ensemble3.eval()

# 12. Hàm refine: xác định lưỡng lự và phân loại lại
def is_ambiguous_3class(probs, bkl_idx, nv_idx, mel_idx, threshold=0.2):
    # probs: (batch, 9)
    top4_idx = np.argsort(probs)[::-1][:4]
    top3_idx = np.argsort(probs)[::-1][:3]
    # 1 trong 3 nhãn đứng top 1
    if np.argmax(probs) not in [bkl_idx, nv_idx, mel_idx]:
        return False

    # Cả 3 nhãn đều nằm trong top 4
    if not all(idx in top4_idx for idx in [bkl_idx, nv_idx, mel_idx]):
        return False

    if not all(idx in top3_idx for idx in [nv_idx, mel_idx]):
        return False
        
    # # Hiệu xác suất top 1 với top 2 hoặc 3 (phải là 1 trong 3 nhãn)
    # idx_top1 = np.argmax(probs)
    # top4_3class = [idx for idx in top4_idx if idx in [bkl_idx, nv_idx, mel_idx] and idx != idx_top1]
    # for idx in top4_3class:
    #     diff = abs(probs[idx_top1] - probs[idx])
    #     if diff < threshold:
    #         return True
    return True

# 13. Đánh giá các chiến lược
def evaluate_refine_strategy(
    model_9class, model_vit3, model_eff3, model_ensemble3, test_loader, class_names, meta_feature_cols, 
    refine_type='none', threshold=0.2
):
    # refine_type: 'none', 'vit', 'eff', 'ensemble'
    model_9class.eval()
    model_vit3.eval()
    model_eff3.eval()
    model_ensemble3.eval()
    bkl_idx = class_names.index('BKL') if 'BKL' in class_names else 0
    nv_idx = class_names.index('NV') if 'NV' in class_names else 1
    mel_idx = class_names.index('MEL') if 'MEL' in class_names else 2
    all_preds = []
    all_labels = []
    all_probs = []
    for batch in tqdm(test_loader, desc=f'Evaluate ({refine_type})'):
        if len(batch) == 3:
            images, metadata, targets = batch
            images = images.to(device)
            metadata = metadata.to(device)
            targets = targets.to(device)
        else:
            images, targets = batch
            images = images.to(device)
            metadata = None
            targets = targets.to(device)
        with torch.no_grad():
            outputs_9 = model_9class(images, metadata)
            probs_9 = F.softmax(outputs_9, dim=1).cpu().numpy()
            batch_preds = []
            batch_probs = []
            for i in range(probs_9.shape[0]):
                probs = probs_9[i]
                # Nếu không refine
                if refine_type == 'none':
                    pred = np.argmax(probs)
                    batch_preds.append(pred)
                    batch_probs.append(probs)
                else:
                    # Kiểm tra lưỡng lự
                    if is_ambiguous_3class(probs, bkl_idx, nv_idx, mel_idx, threshold):
                        # Chuẩn bị input cho model 3 lớp
                        img = images[i].unsqueeze(0)
                        meta = metadata[i].unsqueeze(0) if metadata is not None else None
                        if refine_type == 'vit':
                            out3 = model_vit3(img, meta)
                        elif refine_type == 'eff':
                            out3 = model_eff3(img, meta)
                        elif refine_type == 'ensemble':
                            out3 = model_ensemble3(img, meta)
                        else:
                            raise ValueError("refine_type phải là 'none', 'vit', 'eff', 'ensemble'")
                        prob3 = F.softmax(out3, dim=1).cpu().numpy()[0]
                        # Gán lại xác suất cho 3 lớp, các lớp còn lại = 0
                        new_probs = np.zeros_like(probs)
                        new_probs[bkl_idx] = prob3[0]
                        new_probs[mel_idx] = prob3[1]
                        new_probs[nv_idx] = prob3[2]
                        pred = np.argmax(new_probs)
                        batch_preds.append(pred)
                        batch_probs.append(new_probs)
                    else:
                        pred = np.argmax(probs)
                        batch_preds.append(pred)
                        batch_probs.append(probs)
            all_preds.append(np.array(batch_preds))
            all_probs.append(np.array(batch_probs))
            all_labels.append(targets.cpu().numpy())
    all_preds = np.concatenate(all_preds)
    all_probs = np.vstack(all_probs)
    all_labels = np.concatenate(all_labels)
    # Metrics
    accuracy = accuracy_score(all_labels, all_preds)
    balanced_acc = balanced_accuracy_score(all_labels, all_preds)
    conf_matrix = confusion_matrix(all_labels, all_preds)
    report = classification_report(all_labels, all_preds, target_names=class_names, output_dict=True)
    try:
        y_true_onehot = np.eye(len(class_names))[all_labels]
        auc = roc_auc_score(y_true_onehot, all_probs, multi_class='ovr')
    except Exception as e:
        print(f"Không thể tính ROC AUC: {e}")
        auc = 0.0
    metrics = {
        'accuracy': accuracy,
        'balanced_accuracy': balanced_acc,
        'auc': auc,
        'confusion_matrix': conf_matrix,
        'report': report,
        'all_preds': all_preds, 'all_labels': all_labels, 'all_probs': all_probs 
    }
    return metrics

# 14. Chạy đánh giá cho 4 trường hợp
results = {}
for refine_type in ['none', 'vit', 'eff', 'ensemble']:
    print(f"\n=== Đánh giá chiến lược: {refine_type.upper()} ===")
    metrics = evaluate_refine_strategy(
        model_9class, model_vit3, model_eff3, model_ensemble3, test_loader, class_names, meta_feature_cols, 
        refine_type=refine_type, threshold=0.2
    )
    results[refine_type] = metrics
    # In kết quả tổng quát
    print(f"Accuracy: {metrics['accuracy']:.4f}, Balanced Acc: {metrics['balanced_accuracy']:.4f}, AUC: {metrics['auc']:.4f}")
    print("Macro F1:", metrics['report']['macro avg']['f1-score'])
    print("Confusion matrix:\n", metrics['confusion_matrix'])

# 15. In bảng tổng hợp
print("\n=== BẢNG SO SÁNH HIỆU SUẤT ===")
print(f"{'Chiến lược':<12} {'Acc':<8} {'BalAcc':<8} {'AUC':<8} {'MacroF1':<8}")
for k, v in results.items():
    print(f"{k:<12} {v['accuracy']:<8.4f} {v['balanced_accuracy']:<8.4f} {v['auc']:<8.4f} {v['report']['macro avg']['f1-score']:<8.4f}")

# 16. In báo cáo chi tiết cho từng chiến lược
for k, v in results.items():
    print(f"\n=== {k.upper()} ===")
    print("Classification Report:")
    for cls in class_names:
        cls_metrics = v['report'][cls]
        print(f"{cls:<10} Precision: {cls_metrics['precision']:.4f} Recall: {cls_metrics['recall']:.4f} F1: {cls_metrics['f1-score']:.4f} Support: {cls_metrics['support']}")
    print(f"Macro avg F1: {v['report']['macro avg']['f1-score']:.4f}")

# 17. Vẽ ma trận nhầm lẫn cho từng chiến lược
for k, v in results.items():
    plt.figure(figsize=(8,6))
    cm = v['confusion_matrix']
    cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    sns.heatmap(cm_norm, annot=cm, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.title(f'Confusion Matrix - {k.upper()}')
    plt.ylabel('True')
    plt.xlabel('Pred')
    plt.tight_layout()
    plt.savefig(f'results/confusion_matrix_{k}.png')
    plt.show()
```

```python
def advanced_ensemble_strategy(
    model_9class, model_ensemble3, test_loader, class_names, meta_feature_cols, 
    alpha=0.2, threshold=0.2):
    model_9class.eval()
    model_ensemble3.eval()
    bkl_idx = class_names.index('BKL')
    nv_idx = class_names.index('NV')
    mel_idx = class_names.index('MEL')
    all_preds = []
    all_labels = []
    all_probs = []
    for batch in tqdm(test_loader, desc='Advanced Ensemble'):
        if len(batch) == 3:
            images, metadata, targets = batch
            images = images.to(device)
            metadata = metadata.to(device)
            targets = targets.to(device)
        else:
            images, targets = batch
            images = images.to(device)
            metadata = None
            targets = targets.to(device)
        with torch.no_grad():
            outputs_9 = model_9class(images, metadata)
            probs_9 = F.softmax(outputs_9, dim=1).cpu().numpy()
            batch_size = probs_9.shape[0]
            final_probs = np.copy(probs_9)
            for i in range(batch_size):
                p9 = probs_9[i]
                top4_idx = np.argsort(p9)[::-1][:4]
                top3_idx = np.argsort(p9)[::-1][:3]
                # Điều kiện lưỡng lự nâng cao
                if ((np.argmax(p9) in [bkl_idx, nv_idx, mel_idx]) and all(idx in top4_idx for idx in [bkl_idx, nv_idx, mel_idx])) or ((np.argmax(p9) in [nv_idx, mel_idx]) and all(idx in top3_idx for idx in [nv_idx, mel_idx])):
                    img = images[i].unsqueeze(0)
                    meta = metadata[i].unsqueeze(0) if metadata is not None else None
                    out3 = model_ensemble3(img, meta)
                    p3 = F.softmax(out3, dim=1).cpu().numpy()[0]
                    # Trộn xác suất cho 3 nhãn
                    final_probs[i, bkl_idx] = alpha * p9[bkl_idx] + (1 - alpha) * p3[0]
                    final_probs[i, mel_idx] = alpha * p9[mel_idx] + (1 - alpha) * p3[1]
                    final_probs[i, nv_idx]  = alpha * p9[nv_idx]  + (1 - alpha) * p3[2]
                    # 6 nhãn còn lại giữ nguyên xác suất từ 9 nhãn
            preds = np.argmax(final_probs, axis=1)
            all_preds.append(preds)
            all_probs.append(final_probs)
            all_labels.append(targets.cpu().numpy())
    all_preds = np.concatenate(all_preds)
    all_probs = np.vstack(all_probs)
    all_labels = np.concatenate(all_labels)
    # Metrics
    accuracy = accuracy_score(all_labels, all_preds)
    balanced_acc = balanced_accuracy_score(all_labels, all_preds)
    conf_matrix = confusion_matrix(all_labels, all_preds)
    report = classification_report(all_labels, all_preds, target_names=class_names, output_dict=True)
    try:
        y_true_onehot = np.eye(len(class_names))[all_labels]
        auc = roc_auc_score(y_true_onehot, all_probs, multi_class='ovr')
    except Exception as e:
        print(f"Không thể tính ROC AUC: {e}")
        auc = 0.0
    metrics = {
        'accuracy': accuracy,
        'balanced_accuracy': balanced_acc,
        'auc': auc,
        'confusion_matrix': conf_matrix,
        'report': report,
        'all_preds': all_preds,
        'all_labels': all_labels,
        'all_probs': all_probs
    }
    return metrics

# Sử dụng:
metrics_advanced = advanced_ensemble_strategy(
    model_9class, model_ensemble3, test_loader, class_names, meta_feature_cols, 
    alpha=0.1, threshold=0.2
)
print(f"Advanced Ensemble - Accuracy: {metrics_advanced['accuracy']:.4f}, Balanced Acc: {metrics_advanced['balanced_accuracy']:.4f}, AUC: {metrics_advanced['auc']:.4f}")

# In kết quả đánh giá advanced_ensemble_strategy
print("\n=== ĐÁNH GIÁ ADVANCED ENSEMBLE ===")
print(f"Accuracy: {metrics_advanced['accuracy']:.4f}")
print(f"Balanced Accuracy: {metrics_advanced['balanced_accuracy']:.4f}")
print(f"AUC: {metrics_advanced['auc']:.4f}")
print("Macro F1:", metrics_advanced['report']['macro avg']['f1-score'])

print("\nClassification Report:")
for cls in class_names:
    cls_metrics = metrics_advanced['report'][cls]
    print(f"{cls:<10} Precision: {cls_metrics['precision']:.4f} Recall: {cls_metrics['recall']:.4f} F1: {cls_metrics['f1-score']:.4f} Support: {cls_metrics['support']}")

# Vẽ ma trận nhầm lẫn
plt.figure(figsize=(8,6))
cm = metrics_advanced['confusion_matrix']
cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
sns.heatmap(cm_norm, annot=cm, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix - Advanced Ensemble')
plt.ylabel('True')
plt.xlabel('Pred')
plt.tight_layout()
plt.savefig('results/confusion_matrix_advanced_ensemble.png')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier

# Chuẩn bị dữ liệu validation (val_loader) tương tự test_loader
val_df = pd.read_csv(VAL_LABELS_PATH)
val_df, _, _ = convert_to_multiclass(val_df)

val_dataset = DermDataset(
    img_dir=VAL_IMG_DIR,
    df=val_df,
    metadata_df=metadata_processed,
    meta_cols=meta_feature_cols,
    seg_dir=VAL_SEG_DIR,
    transform=get_valid_transforms(use_seg=True),
    use_seg=True,
    use_meta=True
)
val_loader = torch.utils.data.DataLoader(
    val_dataset, batch_size=32, shuffle=False, num_workers=2, pin_memory=True
)

def train_meta_classifier(val_loader, model_9class, model_ensemble3, class_names, meta_feature_cols):
    model_9class.eval()
    model_ensemble3.eval()
    X_meta = []
    y_meta = []
    for batch in tqdm(val_loader, desc='Meta-train'):
        if len(batch) == 3:
            images, metadata, targets = batch
            images = images.to(device)
            metadata = metadata.to(device)
        else:
            images, targets = batch
            metadata = None
            images = images.to(device)
        with torch.no_grad():
            out9 = model_9class(images, metadata)
            probs9 = F.softmax(out9, dim=1).cpu().numpy()
            out3 = model_ensemble3(images, metadata)
            probs3 = F.softmax(out3, dim=1).cpu().numpy()
        # Ghép xác suất 9 nhãn + 3 nhãn + meta (nếu có)
        for i in range(len(targets)):
            features = list(probs9[i]) + list(probs3[i])
            if metadata is not None:
                features += list(metadata[i].cpu().numpy())
            X_meta.append(features)
            y_meta.append(targets[i].item())
    X_meta = np.array(X_meta)
    y_meta = np.array(y_meta)
    # Bạn có thể dùng LogisticRegression hoặc MLPClassifier
    meta_clf = LogisticRegression(max_iter=1000, multi_class='multinomial')
    meta_clf.fit(X_meta, y_meta)
    return meta_clf

def advanced_ensemble_with_meta(
    model_9class, model_ensemble3, meta_clf, test_loader, class_names, meta_feature_cols,
    alpha=0.2, threshold=0.2, top1_thresh=0.5, topdiff_thresh=0.35):
    model_9class.eval()
    model_ensemble3.eval()
    bkl_idx = class_names.index('BKL')
    nv_idx = class_names.index('NV')
    mel_idx = class_names.index('MEL')
    all_preds = []
    all_labels = []
    all_probs = []
    for batch in tqdm(test_loader, desc='Advanced Ensemble + Meta'):
        if len(batch) == 3:
            images, metadata, targets = batch
            images = images.to(device)
            metadata = metadata.to(device)
            targets = targets.to(device)
        else:
            images, targets = batch
            images = images.to(device)
            metadata = None
            targets = targets.to(device)
        with torch.no_grad():
            out9 = model_9class(images, metadata)
            probs9 = F.softmax(out9, dim=1).cpu().numpy()
            out3 = model_ensemble3(images, metadata)
            probs3 = F.softmax(out3, dim=1).cpu().numpy()
        batch_size = probs9.shape[0]
        final_probs = np.copy(probs9)
        for i in range(batch_size):
            p9 = probs9[i]
            p3 = probs3[i]
            top1 = np.max(p9)
            top2 = np.partition(p9.flatten(), -2)[-2]
            top1_idx = np.argmax(p9)
            top4_idx = np.argsort(p9)[::-1][:4]
            top3_idx = np.argsort(p9)[::-1][:3]
            top2_idx = np.argsort(p9)[::-1][:2]
            # Điều kiện refine nâng cao
            refine = False
            # if np.argmax(p9) in [bkl_idx, nv_idx, mel_idx]:
            #     if top1 < top1_thresh or abs(top1 - top2) < topdiff_thresh:
            #         refine = True
            if (top1 < top1_thresh) and any(idx in top2_idx for idx in [bkl_idx, nv_idx, mel_idx]) and abs(top1 - top2) < topdiff_thresh:
                refine = True
            if ((np.argmax(p9) in [bkl_idx, nv_idx, mel_idx]) and all(idx in top4_idx for idx in [bkl_idx, nv_idx, mel_idx])) or ((np.argmax(p9) in [nv_idx, mel_idx]) and all(idx in top3_idx for idx in [nv_idx, mel_idx])):
                refine = True
            if refine:
                # Trộn xác suất (alpha-blending)
                for idx, p3_idx in zip([bkl_idx, mel_idx, nv_idx], range(3)):
                    final_probs[i, idx] = alpha * p9[idx] + (1 - alpha) * p3[p3_idx]
                # Voting (nếu muốn): pred = mode([np.argmax(p9), np.argmax(p3)])
                # Hoặc stacking meta-classifier:
                features = list(p9) + list(p3)
                if metadata is not None:
                    features += list(metadata[i].cpu().numpy())
                meta_pred = meta_clf.predict([features])[0]
                final_probs[i, :] = 0
                final_probs[i, meta_pred] = 1.0  # one-hot theo meta-classifier
        preds = np.argmax(final_probs, axis=1)
        all_preds.append(preds)
        all_probs.append(final_probs)
        all_labels.append(targets.cpu().numpy())
    all_preds = np.concatenate(all_preds)
    all_probs = np.vstack(all_probs)
    all_labels = np.concatenate(all_labels)
    # Metrics
    accuracy = accuracy_score(all_labels, all_preds)
    balanced_acc = balanced_accuracy_score(all_labels, all_preds)
    conf_matrix = confusion_matrix(all_labels, all_preds)
    report = classification_report(all_labels, all_preds, target_names=class_names, output_dict=True)
    try:
        y_true_onehot = np.eye(len(class_names))[all_labels]
        auc = roc_auc_score(y_true_onehot, all_probs, multi_class='ovr')
    except Exception as e:
        print(f"Không thể tính ROC AUC: {e}")
        auc = 0.0
    metrics = {
        'accuracy': accuracy,
        'balanced_accuracy': balanced_acc,
        'auc': auc,
        'confusion_matrix': conf_matrix,
        'report': report,
        'all_preds': all_preds,
        'all_labels': all_labels,
        'all_probs': all_probs
    }
    return metrics

# Huấn luyện meta-classifier trên tập validation
meta_clf = train_meta_classifier(val_loader, model_9class, model_ensemble3, class_names, meta_feature_cols)

# Đánh giá trên tập test
metrics_meta = advanced_ensemble_with_meta(
    model_9class, model_ensemble3, meta_clf, test_loader, class_names, meta_feature_cols,
    alpha=0.2, threshold=0.2, top1_thresh=0.6, topdiff_thresh=0.2)

import joblib

# Lưu meta-classifier vào file
joblib.dump(meta_clf, 'results/meta_classifier.joblib')
print("Đã lưu meta-classifier vào results/meta_classifier.joblib")

# In kết quả
print("\n=== ĐÁNH GIÁ ENSEMBLE + META-CLASSIFIER ===")
print(f"Accuracy: {metrics_meta['accuracy']:.4f}")
print(f"Balanced Accuracy: {metrics_meta['balanced_accuracy']:.4f}")
print(f"AUC: {metrics_meta['auc']:.4f}")
print("Macro F1:", metrics_meta['report']['macro avg']['f1-score'])

print("\nClassification Report:")
for cls in class_names:
    cls_metrics = metrics_meta['report'][cls]
    print(f"{cls:<10} Precision: {cls_metrics['precision']:.4f} Recall: {cls_metrics['recall']:.4f} F1: {cls_metrics['f1-score']:.4f} Support: {cls_metrics['support']}")

plt.figure(figsize=(8,6))
cm = metrics_meta['confusion_matrix']
cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
sns.heatmap(cm_norm, annot=cm, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix - Advanced Ensemble + Meta')
plt.ylabel('True')
plt.xlabel('Pred')
plt.tight_layout()
plt.savefig('results/confusion_matrix_advanced_ensemble_meta.png')
plt.show()
```
